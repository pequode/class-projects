```python
import numpy as np
from numpy import inf

def vertices(W):
    return set(x[0] for x in W).union(set(x[1] for x in W))


class Vertex():
    def __init__(self,node_id):
        self.name=node_id
        self.d=np.inf
        self.pi=None
    def __str__(self):
        return '{}: d:{} pi:{}'.format(self.name,self.d,self.pi)

class Graph():
    def __init__(self,weights):
        self.V={}
        self.Adj={}
        self.w={}
        if weights:
            for node_id in vertices(weights):
                self.V[node_id]=Vertex(node_id)
                self.Adj[node_id]=[]
            for (st,en),weight in weights.items():
                self.Adj[st].append(en)
                self.w[(st,en)]=weight
                
    def __str__(self):
        vs=sorted(self.V)
        return "\n".join(str(self.V[v]) for v in vs)+"\n"+"\n".join(str(v)+": "+str(self.Adj[v]) for v in vs)

    def initialize_source(self,source,delay=0):
        for v in self.V:
            self.V[v].d = inf
            self.V[v].pi = None
        
        self.V[source].d = -delay
def relax(u,v,G):
    if G.V[v].d > G.V[u].d + G.w[u,v]:
        G.V[v].d = G.V[u].d + G.w[u,v]
        G.V[v].pi = u

def Dijkstra(G,s):
    S=set()
    G.initialize_source(s)
    Q=set(G.V)
    while Q:
        u = min(Q,key=lambda x: G.V[x].d)
        Q.remove(u)
        S.add(u)
        for v in G.Adj[u]:
            relax(u,v,G)


def get_path(G,s,v):
    if s == v:
        return [s]
    elif not G.V[v].pi:
        return None
    else:
        return get_path(G,s,G.V[v].pi) + [v]

def numberize_vertices(W):
    WN={}
    V=sorted(vertices(W))
    Vmap={V[i]:i for i in range(len(V))}

    for edge in W:
        src,dest=edge
        nsrc,ndest=Vmap[src],Vmap[dest]
        WN[(nsrc,ndest)]=W[edge]
    return WN,V



if __name__ == '__main__':
    #name links here with waits 
    W={'AB':1,'AH':4,
       'BC':2,'BI':2,
       'CD':1,
       'DE':2,'DK':4,
       'EF':1,
       'FG':2,'FM':1,
       'HI':4,'HO':3,
       'IJ':2,'IP':1,
       'JK':3,'JQ':4,
       'KL':4,
       'LM':3,'LS':3,
       'MN':2,'MT':1,
       'NU':3,
       'OP':1,'OV':1,
       'PW':1,'PQ':1,
       'QX':1,'QR':4,
       'RY':1,'RS':1,
       'SZ':2,'ST':3,
       'Ta':3,'TU':3,
       'Ub':4,
       'VW':2,
       'WX':4,'Wd':2,
       'Xe':4,'XY':1,
       'Yf':2,
       'Za':4,
       'ab':1,
       'bi':4,
       'cj':2,'cd':3,
       'el':2,
       'dk':2,'de':1,
       'fm':4,'fg':3,
       'gh':3,
       'ho':3,'hi':4,
       'ip':3,
       'jq':2,
       'kr':1,
       'ls':1,'lm':1,
       'mt':3,'mn':4,
       'no':2,
       'ov':3,
       'pw':1,
       'qr':2,
       'tu':2,
       'uv':3 
        }
    NodeNames=tuple('ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvw')
    G=Graph(W)
    Dijkstra(G,'q')
    print(G)
    print(get_path(G,'G','q'))
    WN,VV=numberize_vertices(W)
    print(WN)
    print(W)
    print(VV)
```
