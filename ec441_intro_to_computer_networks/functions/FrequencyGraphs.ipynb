```python
T=0.2 #symbol time
dt=0.01
S=5
t=r_[-S:S:dt]

g = (t>-T/2) * (t<T/2)
#g=cos(2*pi*2*t)
```

```python
plot(t,g)
```

```python
G=fft.fft(g)*dt # G(f) is the Fourier Transform of the pulse shape g(t)
G=fft.fftshift(G)
MaxF = 0.5/dt
df= 1/(2*S)
f = r_[-MaxF:MaxF:df]
figure(figsize=(12,4))
subplot(121)
plot(f,abs(G))
subplot(122)
plot(f,abs(G))
xlim(-20,20);
```
