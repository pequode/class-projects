```python
from scipy.special import erfc
```

```python
Q=lambda x: 0.5*erfc(x/(2**0.5))
```

```python
EbNo = linspace(0.01,100,1000)
Pe = Q((2*EbNo)**0.5)
```

```python
figure(figsize=(8,8))

semilogy(10*log10(EbNo),Pe)
grid()
xlabel('Eb/No (dB)')
ylabel('Probability of Error')
xlim(0,15)
ylim(10e-16,10e-1)
```

```python
N=4
M=2**N
symbols=[2*i-M+1 for i in range(M)]
print(symbols)
```

```python
Power = sum([x**2 for x in symbols])/M
print(Power)
```
