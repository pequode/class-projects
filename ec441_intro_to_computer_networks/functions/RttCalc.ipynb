```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from numpy.random import choice


alpha = 0.125
beta = 0.05

fig, ax = plt.subplots()
#time, RTTs, EstRTTs,DevRTTs,Timeout = [1], [50], [50], [16], [114]
time, RTTs, EstRTTs,DevRTTs,Timeout = [], [], [], [], []

ln_rtt, = plt.plot([], [], 'ro',markersize=2)

ln_estrtt, = plt.plot([], [], 'g')

ln_devrtt, = plt.plot([],[],"b")
ln_timeout, = plt.plot([],[],"y")

ax.set_title(f"RTT estimation ($\\alpha$={alpha})")
ax.set_ylabel("Round Trip Time (ms)")
ax.set_xlabel("RTT count (last 100)")
ax.legend(("Sample","EstRTT","DevRTT","Timeout"))
ax.grid()

def get_rtt():
  return choice([40,90],p=[0.8,0.2])

def init():
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 150)
    return ln_rtt,

def update(frame):
    time.append(frame)
    SampleRTT=get_rtt()
    RTTs.append(SampleRTT)
    if EstRTTs:
     EstRTTs.append((1-alpha)*EstRTTs[-1] + alpha*SampleRTT)
    else:
      EstRTTs.append(SampleRTT)

    if DevRTTs:
        DevRTTs.append((1-beta)*DevRTTs[-1] + beta*abs(SampleRTT-EstRTTs[-1]))
    else:
        DevRTTs.append(0)

    Timeout.append(EstRTTs[-1]+4*DevRTTs[-1])
    xaxis=time[-100:]

    ln_rtt.set_data(xaxis, RTTs[-100:])
    ln_estrtt.set_data(xaxis,EstRTTs[-100:])
    ln_devrtt.set_data(xaxis,DevRTTs[-100:])
    ln_timeout.set_data(xaxis,Timeout[-100:])

    ax.set_xlim(xaxis[0],max(time[-1],100))
    # ax.set_xticks([xaxis[0],xaxis[len(xaxis)//2],xaxis[-1]])
    # ax.set_xticklabels([xaxis[0],xaxis[len(xaxis)//2],xaxis[-1]])
    #ax.set_xbound(lower=xaxis[0],upper=xaxis[-1])
    return (ln_rtt,ln_estrtt,ln_devrtt,ln_timeout)

ani = FuncAnimation(fig, update, frames=np.r_[1:1000],
                    init_func=init, blit=True,interval=40,repeat=False)

plt.show()
print("Average Timeout",np.mean(Timeout))
print("Average deviation RTT",np.mean(DevRTTs))
print("Average Estimated RTT",np.mean(EstRTTs))
# EstRTTs.append(EstRTT)
```
