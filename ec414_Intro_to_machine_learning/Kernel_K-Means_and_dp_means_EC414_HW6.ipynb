###**b.** K-means objective (5 points)
Implement the K-means objective (within-cluster sum of squares) function: $\sum\limits_{i=1}^k Var(C_i) = \sum\limits_{i=1}^k \sum\limits_{\mathbf x \in C_i} ||\mathbf x-\mathbf{\mu_i}||_2^2$

Note: we recommend that you either use njit or not use loops here. More note on njit can be found in question 3.b where a loop over the data is inevitable.

```python
#@njit(fastmath=True,cache=True)
def WCSS(X, labels, m):

  (d,k)=m.shape
  clusters= list(range(d))
  normVes = -1*np.ones((k))

  for j in range(d): # small loop 
     x = X[np.nonzero(labels == j)]

     NotNice = np.array(m)[j]
     #print(x.shape)
     #inner = (x*x)+(NotNice *NotNice)-(2*x*NotNice)
     normvec = np.linalg.norm(x-NotNice)
     inner = normvec * normvec
     sm= np.sum(inner)

     if(sm <0):
       print("negitve objective: ",sm)
       quit()
     clusters[j]  = sm

  return (np.sum(np.array(clusters)))

WCSS(DATA, DATA_labels, means)
```

```python
## DP-means main function:
def DP_means(X, m, Lambda=4, maxT=10, threshold=.001, plots=False):
    (d, k), n = m.shape, X.shape[0]
    labels = np.zeros((n,1))
    # WRITE CODE HERE:
    last = -1000
    obj = []
    mz = m
    for t in range(maxT):
        (d, k)= mz.shape
        # 1. Implement the onepass function and use it here
        if int(plots) == 1: scatter_by_category(X, labels, mz.T)
        mz,lables = onepass(X, mz.T, Lambda, labels)
        # 2. Conditionally plot clustering results:
        
        # 3. Mean updating - Update the cluster means
        # Similar to K-means, be careful about empty clusters
        for j in range(k):# small loop 
          if (j in list(labels)):
            l = X[np.nonzero(labels == j)[0]]
            newM = np.mean(l,axis =0).T
            mz[:,j] =  newM
            
        # 4. Check for convergence based on how much k and m changed
        obj.append(DP_means_obj(X, mz, Lambda, labels))
        if np.absolute(obj[t]-last) <= threshold: break
        last = obj[t]

    if (int(plots) == -1):
      plt.plot(obj,'-r')

    return (mz, t+1,obj[t])
```
